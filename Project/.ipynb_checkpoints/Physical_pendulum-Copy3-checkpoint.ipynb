```python
#import nessecery packages
import numpy as np
import math
import matplotlib.pyplot as plt
import csv
```

```python
def realpendulum(g,l,l2,dt,t_0,t_f,t_f2,theta0,theta6,q,FD,omegaD,y_n):
    #creating a module to calculate the motion of a pendulum with parameters g=gravity,l=length to blob, dt= timestep size, t_0= first timestep
    #t_f= how many seconds the program must calculate of a realistic pendulum, t_f2= how many seconds the program must calculate of a simple pendulum
    #theta0= starting angle of the realistic pendulum, theta6= starting angle of the simple pendulum, q = damping coefficient,FD = driving force,
    #omegaD = angular frequency of the driving force, y_n = to find whic program to run
    
    #calculating amount of timesteps
    number1 = (t_f-t_0)/dt
    number2 = (t_f2-t_0)/dt
    number = int(number1)
    number2 = int(number2)
    
    #declearing variables, which is used in the calculations w(1,2,3) = angular velocity of the different types of pendulum, theta(,2,3) = different 
    #theta values for different pendulums, t(1,2,3)= the value of time as it moves along, delthe= calculating the difference between theta and theta3
    w1 = [[0 for x in range(number)] for y in range(number)]
    w2 = [[0 for x in range(number2)] for y in range(number2)]
    w3 = [[0 for x in range(number)] for y in range(number)]
    theta = [[0 for x in range(number)] for y in range(number)]
    theta2 = [[0 for x in range(number2)] for y in range(number2)]
    theta3 = [[0 for x in range(number)] for y in range(number)]
    t1 = [[0 for x in range(number)] for y in range(number)]
    t2 = [[0 for x in range(number2)] for y in range(number2)]
    t3 = [[0 for x in range(number)] for y in range(number)]
    delthe = [[0 for x in range(number)] for y in range(number)]
    
    #If statement to see which calculation must be done. y_0 == 0 (Using the Euler Cromer method to calculate a realistic pendulum), y_n ==1
    #(Using Euler method to calculate a simple pendule), y_n == 2 (Using the Verlet method to calculate a simple pendulum)
    #The while loops for j runs for different FD values, thus making it possible to plot different graphs with one calculation. In the 
    #simple pendulum the while loop for j, loops only a single time
    if y_n==0:
        j=0
        while j<3:
            #initializing the values to do the calculation for. The difference in 1 and 3 is just a pendulum starting at a different theta
            #to see how the different pendulums react
            w1[j][0] = 0
            theta[j][0] = theta0
            t1[j][0] = t_0
            w3[j][0] = 0
            theta3[j][0] = theta0 +0.001
            t3[j][0] = t_0
    
            i=0
            #Calculating process using the Euler Cromer method for a Realistic pendulum.
            while i<number1-1:
                w1[j][i+1] = w1[j][i] -(g/l)*np.sin(theta[j][i])*dt - q*w1[j][i]*dt + FD[j]*np.sin(omegaD*t1[j][i])*dt
                w3[j][i+1] = w3[j][i] -(g/l)*np.sin(theta3[j][i])*dt - q*w3[j][i]*dt + FD[j]*np.sin(omegaD*t3[j][i])*dt
                theta[j][i+1] = theta[j][i] +w1[j][i+1]*dt
                theta3[j][i+1] = theta3[j][i] +w3[j][i+1]*dt
                #The if function resets the theta to keep theta within a range of pi from zero. This needs to be commented out if one wants to
                #calculate the difference in theta for two pendulums
                if theta[j][i+1]<(-1*np.pi):
                    thetanew = theta[j][i+1]+2*np.pi
                    theta[j][i+1] = thetanew
                elif theta[j][i+1]> 1*np.pi:
                    thetanew1 = theta[j][i+1]-2*np.pi
                    theta[j][i+1] = thetanew1
        
                t1[j][i+1] = t1[j][i] + dt
                t3[j][i+1] = t3[j][i] + dt
                #Calculaing the difference between two pendulums
                delthe[j][i+1] = np.absolute(theta[j][i+1] - theta3[j][i+1])
                i = i+1
            j = j+1
            
    elif y_n ==1:
        j=0
        while j<1:
            #initializing the varaibles that is going to be used to calculate a simple pendulum
            theta2[j][0] = theta6
            t2[j][0] = t_0
            w2[j][0] = np.sqrt(g/l2)
            i=0
            #Calculating process using the Euler method for a simple pendulum.
            while i<number2-1:
                w2[j][i+1] = w2[j][i] - (g/l2)*theta2[j][i]*dt
                theta2[j][i+1] = theta2[j][i] + w2[j][i]*dt
                t2[j][i+1] = t2[j][i] + dt
                i = i+1
            j=j+1
            
    elif y_n ==2:
        j=0
        while j<1:
            #initializing the varaibles that is going to be used to calculate a simple pendulum
            i=0
            w2[j][0] = np.sqrt(g/l2)
            theta2[j][0] = theta6
            t2[j][0] = t_0
            #initializing the second array value using the Euler method
            w2[j][1] = w2[j][0] - (g/l2)*theta2[j][i]*dt
            theta2[j][1] = theta2[j][0] +w2[j][0]*dt
            t2[j][1] = t2[j][0] + dt
            i=1
            #Calculating process using the Verlet method for a simple pendulum.
            while i<number2-1:
                theta2[j][i+1] = 2*theta2[j][i] -theta2[j][i-1] - (g/l2)*(theta2[j][i])*(dt**2)
                w2[j][i] = 0.5*(theta2[j][i+1]-theta2[j][i-1])/dt
                t2[j][i+1] = t2[j][i] + dt
                i = i+1
            j=j+1
    # returing al the calculated values
    return w1,theta,t1,w2,theta2,t2, delthe,w3,theta3,t3
```

```python
#Asking which calculation methods must be used
y_n = input('Euler cromer = 0, Euler method = 1, verlet = 2')
#Initial values
omegaD = (2/3.0)
FD = np.array([0,0.5,1.2])
q = 0.5
theta =0.2
theta2 = np.radians(10)
tf2 = 10
tf1=60
g=9.8
l=9.8
l2 = 1.0

#Calling the pendulum program with parameters, g=gravity =9.8,l=length to blob=9.8, dt= timestep size=0.04, t_0= first timestep=0
    #t_f= how many seconds the program must calculate of a realistic pendulum=60, t_f2= how many seconds the program must calculate of a simple pendulum
    #=10, theta0= starting angle of the realistic pendulum = 0.2, theta6= starting angle of the simple pendulum= np.radians(10), q = damping coefficient=0.5,FD = driving force,
    #=[0,0.5,1.2], omegaD = angular frequency of the driving force=2/3, y_n = to find which program to run 
w1,theta,t1,w2,theta2,t2,delthe,w3,theta3,t3 = realpendulum(g,l,l2,0.04,0,tf1,tf2,theta,theta2,q,FD,omegaD,y_n)

# start a graph page to plot various calculations on one graph
fig = plt.figure()
if y_n ==0:
    #plotting for the realistic pendulum using Euler Cromer
    f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
    ax1.set_title('theta versus time')
    ax1.plot(t1[0], theta[0], color='m')
    ax2.plot(t1[1], theta[1], color='b')
    ax3.plot(t1[2], theta[2], color='r')
    f.subplots_adjust(hspace=0)
    plt.savefig("Realistic_pendulum_final.pdf")
    
    #uncomment if you want to plot the difference in theta for the two pendulums
    #fig = plt.figure()
    #f, ax = plt.subplots(1)
    #plt.xlabel('time (s)')
    #plt.ylabel('Delta theta (radians)')
    #plt.title('Delta theta vs time')
    #j=1
    #while j<3:
    #    plt.plot(t3[j], delthe[j])
    #    j = j+1
    #plt.yscale('log')
    #plt.grid(True)
    #plt.legend(bbox_to_anchor=(0.15, 0.95), loc=1, borderaxespad=0.)
    #ax.set_xlim(xmin=0)
    #plt.savefig("Realistic_pendulum_delthe.pdf")
    #plt.show()
    
elif y_n==1:
    #plotting for the Simple pendulum using Euler
    f, ax = plt.subplots(1)
    titles= 'Simple pendulum - Euler'
    plt.plot(t2[0], theta2[0],label="length = 1, dt=0.001s",color = 'c')
    plt.xlabel('Time (s)')
    plt.ylabel('Angle (radians)')
    plt.title(titles)
    ax.set_xlim(xmin=0)
    plt.savefig("Simple_pendulum_Euler_final1.pdf")
elif y_n==2:
    #plotting for the Simple pendulum using Verlet
    f, ax = plt.subplots(1)
    titles= 'Simple pendulum - Verlet'
    plt.plot(t2[0], theta2[0],label="length = 1, dt=0.04s", color = 'm')
    plt.xlabel('Time (s)')
    plt.ylabel('Angle (radians)')
    plt.title(titles)
    ax.set_xlim(xmin=0)
    plt.savefig("Simple_pendulum_Verlet_final.pdf")

plt.show(fig)

#Saving the data, an x and a y-value per folder. comment out if not needed 
#with open('data_week3_Realistic_pendulum1.txt', 'w') as f:
#    writer = csv.writer(f, delimiter='\t')
#    writer.writerows(zip(t1,theta))
#with open('data_week3_Realistic_pendulum2.txt', 'w') as f:
#    writer = csv.writer(f, delimiter='\t')
#    writer.writerows(zip(t1,theta))
#with open('data_week3_Realistic_pendulum2.txt', 'w') as f:
#    writer = csv.writer(f, delimiter='\t')
#    writer.writerows(zip(t1,theta))
```
