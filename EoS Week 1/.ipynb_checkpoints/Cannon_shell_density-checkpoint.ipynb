```python
#importing importent packages
import numpy as np #numerical python
import matplotlib.pyplot as plt #plotting package
import csv #saving data package
```

```python
#creating module for the cannon ball trajectory, with parameters: number = amount of time steps,x1_0=y1=v1_0= initial values for the coordinates
#and velocity, theta= firing angle, dt= step size, B2m = drag parameter, g= gravity, m= mass, rho0= density of air at sea level and 
#at 15 degree celsius
def CannonS(number,x1_0,y1_0,v_0,theta,dt,B2m,g,m,rho0):
    #declearing the coordinate components,velocity components, and drag components. Its done in this way since i want to create a 
    #multidimensional array 
    x1 = [[0 for x in range(number)] for y in range(number)]
    y1 = [[0 for x in range(number)] for y in range(number)]
    vx1 = [[0 for x in range(number)] for y in range(number)]
    vy1 = [[0 for x in range(number)] for y in range(number)]
    fx = [[0 for x in range(number)] for y in range(number)]
    fy = [[0 for x in range(number)] for y in range(number)]
    
    # creating a value k to store the position the of the number in the array when the ball iss about to hit the ground
    k = np.zeros(5)
    #starting first array j for different theta. for each theta a x and y position will be calculated as the ball moves trough the air
    j=0
    while j<5:
        #initializing values for each theta
        x1[j][0]=x1_0
        y1[j][0]=y1_0
    
        vx1[j][0]=v_0*np.cos(theta[j])
        vy1[j][0]=v_0*np.sin(theta[j])
       
        i=0
        #start computing the x and y position at each time step for a certain theta
        while y1[j][i]>=0 :
            
            #using rho to correct for density verying at different altitudes
            x1[j][i+1]=(x1[j][i]+ (vx1[j][i]*dt))
            y1[j][i+1]=(y1[j][i]+ (vy1[j][i]*dt))
            rho = rho0*np.exp(-(y1[j][i])/(y1_0+10000))
            #calculating the drag force at different altitudes
            fx[j][i]= -(B2m*rho/(rho0))*v_0*vx1[j][i]
            fy[j][i]= -(B2m*rho/(rho0))*v_0*vy1[j][i]

            #calculating the velocity to determine if it is going up or down.
            vx1[j][i+1]=vx1[j][i]+(fx[j][i]/m)*dt
            vy1[j][i+1]=vy1[j][i]+(-g+fy[j][i]/m)*dt
            #saving the position of the last possible i for a + y value
            k[j]=i
            #test if it below flour level. If it is interpolate, to find where the ball landed between y[i] and y[i+1]
            if y1[j][i+1]<=0:
                r = -y1[j][i]/y1[j][i+1]
                x_l= (x1[j][i]+r*x1[j][i+1])/(r+1)
                #saving that in the middle x as the last position in the array
                x1[j][i+1]=x_l
                print 'With density correction at theta=',theta[j]
                print 'The ball hits the ground at =',x1[j][i+1]
                break
            i=i+1
            
        #changing the last value of y to 0 since its on the ground
        y1[j][i+1]=0
        print k[j]
        j=j+1
        
    #doing the same as above but without the density correction
    #while j<4:
    #    x1[j][0]=x1_0
    #    y1[j][0]=y1_0
    
    #    vx1[j][0]=v_0*np.cos(theta[j-2])
    #    vy1[j][0]=v_0*np.sin(theta[j-2])
       
    #    i=0
    #    while y1[j][i]>=0 :
    #        x1[j][i+1]=(x1[j][i]+ (vx1[j][i]*dt))
    #        y1[j][i+1]=(y1[j][i]+ (vy1[j][i]*dt))
        
    #        fx[j][i]= -(B2m)*v_0*vx1[j][i]
    #        fy[j][i]= -(B2m)*v_0*vy1[j][i]
        
    #        vx1[j][i+1]=vx1[j][i]+(fx[j][i]/m)*dt
    #        vy1[j][i+1]=vy1[j][i]+(-g+fy[j][i]/m)*dt
    #        k[j]=i
     #       if y1[j][i+1]<=0:
     #           r = -y1[j][i]/y1[j][i+1] 
     #           x_l= (x1[j][i]+r*x1[j][i+1])/(r+1)
     #           x1[j][i+1]=x_l
     #           print 'Without density correction at theta=',theta[j-2]
     #           print 'The ball hits the ground at = ',x1[j][i+1]
     #           break
     #       i=i+1
            
        
     #   y1[j][i+1]=0
      #  print k[j]
       # j=j+1
        
    return x1,y1,k,number
```

```python
B2m=4*(10**-5)
theta =np.array([np.radians(30),np.radians(40),np.radians(45),np.radians(50),np.radians(60)])
#calling the module for the cannon but with different initial values
x1,y1,k,number = CannonS(150,0,0,700,theta,1,B2m,9.81,40,1.225)
#creating stuff for the graph such as the title name
theta1 =np.array([41,43,45,47,49])
titles= 'Trajectory of cannon shell with drag,different angles and density correction'
j=0
#slicing the array at the last value y[i+1], but since k was saved befor the if at y[i] we need to start slicing only at k+2
while j<5:
    x1[j][:]=x1[j][0:int(k[j]+2)]
    y1[j][:]=y1[j][0:int(k[j]+2)]
    j = j+1

#start plotting the graph
fig = plt.figure()
f, ax = plt.subplots(1)

#graph labels
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title(titles)
j=0

#plot and scatter plot the graph at different thetas with denisity corretion
while j<5:
    plt.plot(x1[j], y1[j],label=(theta1[j]))
    plt.scatter(x1[j],y1[j],s=7)
    j = j+1
    
#on graph text
#plt.text(20000, 4500, 'With density correction', fontdict=None, withdash=False)

#plot and scatter plot the graph at different thetas without density corretion
#while j<4:
#    plt.plot(x1[j], y1[j],label=(theta1[j-2]))
#    plt.scatter(x1[j],y1[j],s=7)
#    j = j+1
    
#on graph text
#plt.text(11000, 1500, 'Without density correction', fontdict=None, withdash=False)

#stuff to make it look nice
plt.grid(True)
plt.legend(bbox_to_anchor=(0.15, 0.95), loc=1, borderaxespad=0.)

#save figure
plt.savefig("trajectory_cannon_density_at different_angles_test4.pdf")
plt.show()
j=0

#Saving the data with each a x and a y value
#while j<5: 
#    label1 = 'data_week2_trajectory_density_firing_test3_%d.txt' %(j)
#    with open(label1, 'w') as f:
#       writer = csv.writer(f, delimiter='\t')
#       writer.writerows(zip(x1[j],y1[j]))
#    j=j+1
```
